```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Sigmoid

```python
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
  
plt.plot(x, z)
plt.xlabel("X")
plt.ylabel("Sigmoid(X)")
  
plt.show()
```

# Tanh

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10, 100)
#z = tanh(x)

plt.plot(x, tanh(x))
plt.xlabel("X")
plt.ylabel("Tanh(X)")
  
plt.show()
```

# RELU

```python
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10, 100)

plt.plot(x, RELU(x))
plt.xlabel("X")
plt.ylabel("Tanh(X)")
  
plt.show()
```

# Softmax

```python
def softmax(x):
    ''' Compute softmax values for each sets of scores in x. '''
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-10, 10, 100)


plt.plot(x, softmax(x))
plt.xlabel("X")
plt.ylabel("Tanh(X)")
  
plt.show()
```
